```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.api import VAR
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
df =pd.read_csv('./Microsoft_stock_data.csv', parse_dates=['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
train_size = int(0.8 * len(df))
train_df, test_df = df.iloc[:train_size], df.iloc[train_size:]
```

```python
scaler_var = StandardScaler()
train_scaled = pd.DataFrame(scaler_var.fit_transform(train_df), columns=df.columns, index=train_df.index)
test_scaled = pd.DataFrame(scaler_var.transform(test_df), columns=df.columns, index=test_df.index)
```

```python
model_var = VAR(train_scaled)
var_fit = model_var.fit(maxlags=5)
print(f"✅ VAR fitted. Selected lag order: {var_fit.k_ar}")
```

```python

# We use the fitted VAR to forecast one-step-ahead across the training period so the LSTM is trained only on in-sample residuals.
k = var_fit.k_ar
train_vals = train_scaled.values
# history will hold the observed values (not forecasts) used to produce the next-step forecast
history = list(train_vals[:k].tolist())
train_forecasts = []
for t in range(k, len(train_vals)):
    # forecast one step ahead using the most recent k values from history
    f = var_fit.forecast(np.array(history[-k:]), steps=1)[0]
    train_forecasts.append(f)
    # append the true next observation to history (walk-forward)
    history.append(train_vals[t])
# convert to DataFrame aligned with train index starting at position k
train_forecasts = np.vstack(train_forecasts) if len(train_forecasts)>0 else np.empty((0, train_scaled.shape[1]))
train_forecast_df = pd.DataFrame(train_forecasts, index=train_scaled.index[k:], columns=train_scaled.columns)
train_residuals = train_scaled.iloc[k:].copy() - train_forecast_df
# drop any NaNs that might remain
train_residuals = train_residuals.dropna()
print(f"✅ Train residuals generated. Shape: {train_residuals.shape}")
```

```python
steps = len(test_scaled)
forecast_values = var_fit.forecast(train_scaled.values[-var_fit.k_ar:], steps=steps)
forecast_df = pd.DataFrame(forecast_values, index=test_scaled.index, columns=test_scaled.columns)
print(f"✅ Test forecasts generated. Shape: {forecast_df.shape}")
```

```python
# Compute test residuals (do not fill with zeros to avoid bias).
residuals = test_scaled - forecast_df
# drop rows with NaNs if any (safer than filling with zeros)
residuals = residuals.dropna()
print(f"✅ Test residuals computed. Shape: {residuals.shape}")
```

```python
# Fit the LSTM scaler on in-sample (train) residuals, then transform test residuals
scaler_lstm = MinMaxScaler()
# train_residuals was generated earlier (walk-forward). Fit scaler on those in-sample residuals
train_residuals_scaled = pd.DataFrame(scaler_lstm.fit_transform(train_residuals), columns=train_residuals.columns, index=train_residuals.index)
# transform the test residuals using the same scaler (do not fit again)
residuals_scaled_test = pd.DataFrame(scaler_lstm.transform(residuals), columns=residuals.columns, index=residuals.index)
print(f"✅ Scaler for LSTM fitted on train residuals. Train scaled shape: {train_residuals_scaled.shape}, Test scaled shape: {residuals_scaled_test.shape}")
```

```python
def create_sequences(X, seq_len=10):
    Xs, ys = [], []
    for i in range(len(X) - seq_len):
        Xs.append(X.iloc[i:i+seq_len].values)
        ys.append(X.iloc[i+seq_len].values)
    return np.array(Xs), np.array(ys)
```

```python
seq_len = 10
# Create sequences from in-sample (train) residuals for training the LSTM
X_train, y_train = create_sequences(train_residuals_scaled, seq_len)
# Create sequences from test residuals for evaluation
X_test, y_test = create_sequences(residuals_scaled_test, seq_len)
print(f"✅ LSTM data shapes: X_train={X_train.shape}, X_test={X_test.shape}")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# LSTM (using default activations for LSTM cells)
lstm = Sequential([
    LSTM(64, input_shape=(seq_len, X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(32),
    Dense(X_train.shape[2])
])
lstm.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
history = lstm.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1, shuffle=False, callbacks=[es], verbose=1)
```

```python
y_pred = lstm.predict(X_test)
mse = np.mean(np.power(y_test - y_pred, 2), axis=1)
```

```python
iso = IsolationForest(contamination=0.05, random_state=42)
labels = iso.fit_predict(mse.reshape(-1, 1))
anomalies = np.where(labels == -1, 1, 0)
```

```python
# Map results to the appropriate test residual timestamps
res_index = residuals_scaled_test.iloc[-len(mse):].index
results = pd.DataFrame({
    "Reconstruction_Error": mse,
    "Anomaly": anomalies
}, index=res_index)
print(f"✅ Results assembled. Rows: {len(results)}")
```

```python
# Convert to arrays to avoid matrix operations
rmse = np.sqrt(mean_squared_error(np.asarray(y_test).flatten(), np.asarray(y_pred).flatten()))
print(f"\n✅ Hybrid VAR + LSTM RMSE: {rmse:.4f}")

# Create the plot using arrays
close_prices = np.asarray(df["Close"].iloc[-len(results):])
dates = df.index[-len(results):]
anomaly_indices = np.where(results["Anomaly"] == 1)[0]

plt.figure(figsize=(14,6))
plt.plot(dates, close_prices, label="Actual Close", linewidth=1.5)
plt.scatter(dates[anomaly_indices],
            close_prices[anomaly_indices],
            color='red', s=60, label="Anomaly")
plt.title("VAR + LSTM Hybrid Anomaly Detection (Dual Scalers)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

```python
# Filter the original dataframe to show the anomalous data points
anomalous_data = df.loc[results[results['Anomaly'] == 1].index]

print("\n✅ Anomalous Data Points:")
display(anomalous_data)
```

```python
from sklearn.metrics import mean_squared_error

# Collect results
comparison = {
    "VAR+LSTM": {"errors": mse, "anomalies": results["Anomaly"]},

}

records = []
for name, data in comparison.items():
    rmse = np.sqrt(np.mean(np.square(data["errors"])))
    var = np.var(data["errors"])
    count = np.sum(data["anomalies"] == 1)
    records.append({
        "Model": name,
        "Anomalies Detected": count,
        "RMSE": rmse,
        "Residual Variance": var
    })

df_compare = pd.DataFrame(records)
print(df_compare)
```
